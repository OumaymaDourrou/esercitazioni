```python
def posneg(x):
  if x<0:
    return "il num è neg"
  elif x==0:
    return 'il numero inserito è 0, cambiare num'
  else:
    return'il num è pos'
```

```python
posneg(1)
```

```python
x = float(input("inserire un numero: "))

if x<0:
    print ("il num è neg")
elif x==0:
    print ('il numero inserito è 0, cambiare num')
else:
    print('il num è pos')
```

##pari dispari con la formula
